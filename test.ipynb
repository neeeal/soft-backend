```python
from PIL import Image
import base64
from io import BytesIO
import random
import os

def convert_image_to_base64(image_path):
    # Open the image file
    with Image.open(image_path) as img:
        # Convert the image to RGB mode
        rgb_image = img.convert('RGB')
        
        # Save the RGB image to a BytesIO object
        buffered = BytesIO()
        rgb_image.save(buffered, format="JPEG")
        
        # Encode the BytesIO object to base64
        img_base64 = base64.b64encode(buffered.getvalue()).decode("utf-8")
        
    return img_base64

base = 'data/train'
for i in os.listdir(base):
    rel = os.path.join(base,i)
    files = os.listdir(rel)
    num_files = len(files)
    rand_file_idx = random.randint(0,num_files)
    
    img_path = os.path.join(rel,files[rand_file_idx-1])
    base64_image = convert_image_to_base64(img_path)
    print(i)
    print('rice_image: BinData(0, "{}")'.format(base64_image), end="\n\n")
    
# Example usage
# image_path = 'assets\BLB (14).png'
# print(base64_image)
```
